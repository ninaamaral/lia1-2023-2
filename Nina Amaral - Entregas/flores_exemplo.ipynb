```python
!pip install fastai
from fastai.vision.all import *
learn = load_learner('/content/drive/MyDrive/LIA/flowers (1).pk')
```

```python
!pip install numpy
```

```python
im = PILImage.create('/content/drive/MyDrive/LIA/exemplos flor/ipe-amarelo-cascudo_2nd.jpg')
im.thumbnail((192,192))
im
```

```python
import numpy as np
oque,_,probs = learn.predict(im)
print(f"This is a: {oque}.")
w = np.argmax(probs)
print(f"com {probs[w]:.4f} de certeza")
```
